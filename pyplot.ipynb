```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.random.normal(5, 1.5, size=1000)

# Ejemplo con varias opciones de personalización
ax.hist(x, bins=20, color='skyblue', edgecolor='black', alpha=0.7, label='Datos Normales')

# Añadir leyenda
ax.legend()

# Añadir títulos y etiquetas
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma Personalizado')

plt.show()
```

color: Define el color de las barras del histograma. Puedes usar nombres de colores, códigos hexadecimales, etc.
bins: Determina el número de "contenedores" o rangos en los que se agruparán los datos. Puedes especificar un número entero o una secuencia de bordes de contenedor.
density: Si se establece en True, el histograma se normaliza para formar una densidad de probabilidad, es decir, el área bajo el histograma integra a 1.
alpha: Controla la transparencia de las barras (un valor entre 0 y 1).
edgecolor: Define el color del borde de las barras.
linewidth: Establece el ancho de la línea del borde de las barras.
histtype: Define el tipo de histograma a dibujar. Los opciones comunes son 'bar' (el predeterminado), 'barstacked', 'step' y 'stepfilled'.
label: Asigna una etiqueta al histograma para usar en una leyenda.
orientation: Puede ser 'vertical' (predeterminado) u 'horizontal' para dibujar un histograma rotado.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sizes = [5, 4, 3, 2, 1]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.set_title('Ejemplo de Gráfico de Pastel Personalizado')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ingresar datos de 5 personas
nombres = []
pesos = []

for i in range(5):
    nombre = input(f"Ingrese el nombre de la persona {i+1}: ")
    peso = float(input(f"Ingrese el peso de {nombre} (kg): "))
    nombres.append(nombre)
    pesos.append(peso)

# Crear DataFrame
df = pd.DataFrame({
    'Nombre': nombres,
    'Peso (kg)': pesos
})

# Mostrar datos
print("\nDatos ingresados:")
print(df)

# Calcular media y desviación estándar
media = df['Peso (kg)'].mean()
desviacion = df['Peso (kg)'].std()

print(f"\nMedia de peso: {media:.2f} kg")
print(f"Desviación estándar: {desviacion:.2f} kg")

# Definir colores para cada sector
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']

# Definir explode para destacar el primer sector (puedes ajustar)
explode = [0.1] + [0]*(len(df)-1)

# Graficar diagrama de pastel con estilo
fig, ax = plt.subplots()
ax.pie(
    df['Peso (kg)'],
    explode=explode,
    labels=df['Nombre'],
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.2}
)

ax.axis('equal')  # Para que el pastel sea un círculo perfecto
ax.set_title('Distribución de peso por persona')

plt.show()
5
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2, sharey = True)
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax[0, 0].plot(dias, temperaturas['Madrid'])
ax[0, 1].plot(dias, temperaturas['Barcelona'], color = 'tab:orange')
ax[1, 0].bar(dias, temperaturas['Madrid'])
ax[1, 1].bar(dias, temperaturas['Barcelona'], color = 'tab:orange')
plt.show()
```
